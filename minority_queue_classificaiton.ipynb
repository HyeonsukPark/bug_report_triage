# Data Pre-processing

```python
# function to do text pre-processing

def text_cleaning(text):

    text = str(text).lower()

    # Remove text inside square brackets
    text = re.sub(r'\[.*?\]', '', text)

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove words containing numbers (using word boundaries)
    text = re.sub(r'\b\w*\d\w*\b', '', text)

    # Remove any remaining non-alphabetic characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra whitespace (spaces, line breaks)
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
def bug_processing(series):

    stop_words = set(stopwords.words('english'))

    # Apply text cleaning to the entire series
    cleaned_series = series.apply(text_cleaning)

    # Tokenize and remove stopwords using a concise lambda function
    processed_series = cleaned_series.apply(
        lambda text: ' '.join([token for token in word_tokenize(text) if token not in stop_words])
    )

    return processed_series
```

```python
# Apply text cleaning
df['email_content_clean'] = bug_processing(df['email_content'])
```

# LDA

```python
# create a dictionary and corpus
dictionary = corpora.Dictionary(df['email_content_clean'].apply(word_tokenize))
corpus = [dictionary.doc2bow(word_tokenize(text)) for text in df['email_content_clean']]
```

```python
# Run LDA model
num_topics = 5
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics)
```

```python
# get topic distribution for each document
def get_topic_distribution(lda_model, corpus, num_topics):
    topic_distributions = []
    for doc in corpus:
        # get the topic probabilities for each document
        doc_topics = lda_model.get_document_topics(doc, minimum_probability=0)
        # Extract just the probability values in the correct order
        topic_distributions.append([prob for _, prob in doc_topics])
    # Convert the list of lists to a NumPy array
    return np.array(topic_distributions)

# Get the topic distributions to use as features
topic_features = get_topic_distribution(lda_model, corpus, num_topics)
```

```python
lda_df = pd.DataFrame(topic_features, columns=[f'topic_{i}' for i in range(num_topics)])
df = pd.concat([df, lda_df], axis=1)
```

## OpenAI Embedding

```python
import tiktoken
from dotenv import load_dotenv
import dotenv
import os
from openai import OpenAI
```

```python
dotenv.load_dotenv('/content/drive/MyDrive/ML Project/customer_ticket/openai_api.env')
client = OpenAI(
    # API key
    api_key=os.environ.get("OPENAI_API_KEY")
)
```

```python
embedding_model = "text-embedding-3-small"
embedding_encoding = "cl100k_base"  #  the encoding for text-embedding-3-small
max_tokens = 8000 # max input is 8191

def gpt_embedding(texts, batch_size=4, model=embedding_model, encoding=embedding_encoding, max_tokens_limit=max_tokens):
   # Load the tiktoken encoding for the model once
    encoding = tiktoken.get_encoding(encoding)
    all_embeddings = []

    # Process texts in batches with tqdm progress bar
    for i in tqdm(range(0, len(texts), batch_size), desc="Processing batches"):
        batch_texts = []
        # Process each text in the current batch
        for text in texts[i:i+batch_size]:
            # Clean the text by replacing newline characters with space
            text_clean = text.replace("\n", " ")
            token_ids = encoding.encode(text_clean)
            # Truncate the text if token count exceeds the limit
            if len(token_ids) > max_tokens_limit:
                token_ids = token_ids[:max_tokens_limit]
                text_clean = encoding.decode(token_ids)
            batch_texts.append(text_clean)

         # Call the OpenAI API with the batch of texts
        response = client.embeddings.create(input=batch_texts, model=model)
        # Extract embeddings
        batch_embeddings = [result.embedding for result in response.data]
        all_embeddings.extend(batch_embeddings)

    return all_embeddings
```

# Combining embedding and LDA features
Now, we combine embeddings with LDA features (topics)

```python
# Combine embedding and lda_features
# there is a mismatch in dimensions between embeddings and LDA features. embedding is a sparse matrix, while LDA features is a dense NumPy array.
# We convert the dense LDA features into a sparse matrix before combining them.
# using hstack, and csr_matrix for conversion

from scipy.sparse import hstack, csr_matrix

# convert the dense LDA features to a sparse matrix
X_lda_features_train_sparse = csr_matrix(X_lda_features_train.values)
X_lda_features_test_sparse = csr_matrix(X_lda_features_test.values)

combined_X_train = hstack((gpt_X_train, X_lda_features_train_sparse))
combined_X_test = hstack((gpt_X_test, X_lda_features_test_sparse))
```

```python
# The combined matrices will now have the correct shape
print(f'Combined train shape: {combined_X_train.shape}')
print(f'Combined test shape: {combined_X_test.shape}')
```

# Classification

```python
def evaluate_classifiers(model_name, X_train, X_test, y_train, y_test, embedding_name="Embedding"):

    models = {
        "LinearSVC": LinearSVC(random_state=42),
        "RandomForest": RandomForestClassifier(random_state=42),
        "GradientBoosting": GradientBoostingClassifier(random_state=42)
    }

    if model_name not in models:
        raise ValueError("Invalid model name. Choose from 'svm', 'random_forest', 'gradient_boosting'.")

    model = models[model_name]
    print(f"\n Running {model_name.replace('_', ' ').title()} on {embedding_name} Embeddings\n" + "-"*60)

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
    report = classification_report(y_test, y_pred, zero_division=0)

    print(f"🔹 Accuracy     : {accuracy:.4f}")
    print(f"🔹 Precision    : {precision:.4f}")
    print(f"🔹 Recall       : {recall:.4f}")
    print(f"🔹 F1 Score     : {f1:.4f}")
    print("🔹 Classification Report:\n", report)

    return {
        "model": model_name,
        "embedding": embedding_name,
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1_score": f1,
        "report": report
    }
```
